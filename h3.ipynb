```python
'''
1. Modify the Eulerâ€™s method code so you can generate a single plot for the following initial conditions: x0 =
1.0, 2.0, 3.0, 10, 50, 100. Your plot should look like this one:
'''
import matplotlib.pyplot as plt

def euler_solver(func, y0, t0, t_final, delta_t):

    t_values = [t0]
    y_values = [y0]
    t = t0
    y = y0

    while t < t_final:
        y += delta_t * func(t, y)
        t += delta_t

        t_values.append(t)
        y_values.append(y)

    return t_values, y_values

def kcrystal(t, x):
    k = 0.1  # Decay constant
    return -k * x

# Initial conditions for x0 = 1.0, 2.0, 3.0, 10, 50, 100
x0_values = [1.0, 2.0, 3.0, 10, 50, 100]
t0 = 0.0
t_final = 10.0
delta_t = 0.1

# Plotting
plt.figure(figsize=(10, 6))

for x0 in x0_values:
    t_values, x_values = euler_solver(func=kcrystal, y0=x0, t0=t0, t_final=t_final, delta_t=delta_t)
    plt.plot(t_values, x_values, label=f'x0 = {x0}')

plt.xlabel('Time')
plt.ylabel('x(t)')
plt.title('Approximate Solution of dx/dt = -kx using Euler\'s Method')
plt.legend()
plt.grid(True)
plt.show()



```

```python

import matplotlib.pyplot as plt

def euler_solver(func, y0, t0, t_final, delta_t):

    t_values = [t0]
    y_values = [y0]
    t = t0
    y = y0

    while t < t_final:
        y += delta_t * func(t, y)
        t += delta_t

        t_values.append(t)
        y_values.append(y)

    return t_values, y_values

def f(t, y):
    return (5 - y) * (y + 2)

# Initial conditions
t0 = 0
y0 = 0
t_final = 10
delta_t = 0.5

t_values, y_values = euler_solver(f, y0, t0, t_final, delta_t)

# Plot the results
plt.plot(t_values, y_values)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Euler\'s Method Solution for dy/dt = (5 - y)(y + 2)')
plt.grid(True)
plt.show()
```

To find the analytic solution of the initial value problem $\frac{dy}{dt} = (5 - y)(y + 2)$ with the initial condition $y(0) = 0$, we can separate and integrate the variables. Here's the step-by-step solution:

Separate variables:
\[
\frac{dy}{(5 - y)(y + 2)} = dt
\]

Now, we can integrate both sides. To do this, we perform partial fraction decomposition:

\[
\frac{1}{(5 - y)(y + 2)} = \frac{A}{5 - y} + \frac{B}{y + 2}
\]

Multiply both sides by $(5 - y)(y + 2)$ to solve for $A$ and $B$:

\[
1 = A(y + 2) + B(5 - y)
\]

Now, we can solve for $A$ and $B$:

Setting $y = 5$, we get $1 = 7A$, so $A = \frac{1}{7}$.

Setting $y = -2$, we get $1 = -7B$, so $B = -\frac{1}{7}$.

Now, we can rewrite the integral as:

\[
\int \left(\frac{1}{7(5 - y)} - \frac{1}{7(y + 2)}\right) dy = \int dt
\]

Integrate both sides:

\[
\frac{1}{7} \ln|5 - y| - \frac{1}{7} \ln|y + 2| = t + C
\]

Simplify:

\[
\ln\left(\frac{|5 - y|}{|y + 2|}\right) = 7t + C
\]

Take the exponential of both sides:

\[
\frac{|5 - y|}{|y + 2|} = e^{7t + C}
\]

Now, we can remove the absolute values by considering different cases:

Case 1: $y < -2$

In this case, $|5 - y| = -(5 - y)$ and $|y + 2| = -(y + 2)$:

\[
\frac{-(5 - y)}{-(y + 2)} = e^{7t + C}
\]

Simplify:

\[
\frac{5 - y}{y + 2} = e^{7t + C}
\]

Case 2: $-2 < y < 5$

In this case, $|5 - y| = 5 - y$ and $|y + 2| = -(y + 2)$:

\[
\frac{5 - y}{-(y + 2)} = e^{7t + C}
\]

Simplify:

\[
-\frac{5 - y}{y + 2} = e^{7t + C}
\]

Case 3: $y > 5$

In this case, $|5 - y| = 5 - y$ and $|y + 2| = y + 2$:

\[
\frac{5 - y}{y + 2} = e^{7t + C}
\]

Now, we have three different solutions depending on the value of $y$. These are the analytic solutions in the respective ranges. To find the values of $C$ for each range, we can use the initial condition $y(0) = 0$.

1. For $y < -2$, the solution is:

\[
\frac{5 - y}{y + 2} = e^{7t + \ln\left(\frac{5}{2}\right)}
\]

Plugging in $y = 0$ and $t = 0$:

\[
\frac{5}{2} = e^{C}
\]

So, $C = \ln\left(\frac{5}{2}\right)$.

2. For $-2 < y < 5$, the solution is:

\[
-\frac{5 - y}{y + 2} = e^{7t + \ln\left(\frac{5}{2}\right)}
\]

Plugging in $y = 0$ and $t = 0$:

\[
\frac{5}{2} = e^{C}
\]

So, $C = \ln\left(\frac{5}{2}\right)$.

3. For $y > 5$, the solution is:

\[
\frac{5 - y}{y + 2} = e^{7t + \ln\left(\frac{5}{2}\right)}
\]

Plugging in $y = 0$ and $t = 0$:

\[
\frac{5}{2} = e^{C}
\]

So, $C = \ln\left(\frac{5}{2}\right)$.

Now, we have the solutions for each range:

1. For $y < -2$:

\[
\frac{5 - y}{y + 2} = e^{7t + \ln\left(\frac{5}{2}\right)}
\]

2. For $-2 < y < 5$:

\[
-\frac{5 - y}{y + 2} = e^{7t + \ln\left(\frac{5}{2}\right)}
\]

3. For $y > 5$:

\[
\frac{5 - y}{y + 2} = e^{7t + \ln\left(\frac{5}{2}\right)}
\]

These are the analytic solutions in their respective ranges.


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the analytic solution function
def analytic_solution(t):
    C = np.log(5/2)
    return (5 - 2 * np.exp(-7 * t + C)) / (1 + 2 * np.exp(-7 * t + C))

# Define the Euler's method solver (from your previous code)
def euler_solver(func, y0, t0, t_final, delta_t):
    t_values = [t0]
    y_values = [y0]
    t = t0
    y = y0

    while t < t_final:
        y += delta_t * func(t, y)
        t += delta_t

        t_values.append(t)
        y_values.append(y)

    return t_values, y_values

# Initial conditions
t0 = 0
y0 = 0
t_final = 10
delta_t = 0.5

# Solve the ODE using Euler's method
t_values, y_values = euler_solver(f, y0, t0, t_final, delta_t)

# Calculate the analytic solution values at the same time points as Euler's method
analytic_values = [analytic_solution(t) for t in t_values]

# Calculate the absolute error
absolute_error = [abs(y_num - y_analytic) for y_num, y_analytic in zip(y_values, analytic_values)]

# Plot the absolute error
plt.plot(t_values, absolute_error)
plt.xlabel('t')
plt.ylabel('Absolute Error')
plt.title('Absolute Error Between Euler\'s Method and Analytic Solution')
plt.grid(True)
plt.show()
```

4. What are nullclines? Can you spot them for both of the previous systems?

Nullclines are curves in the phase plane (the plane spanned by the variables of the system, typically \(x\) and \(y\)) where one of the derivatives becomes zero. In other words, they are the loci where \(\frac{dx}{dt} = 0\) and/or \(\frac{dy}{dt} = 0\). Nullclines are important in the study of dynamical systems because they represent points where the system's behavior can change, such as equilibrium points and regions where trajectories change direction.

Let's find the nullclines for both of the previous systems:

**System 1:**
1. \(\frac{dx}{dt} = -ax + bxy\)
2. \(\frac{dy}{dt} = cy - dxy\)

To find the nullcline for \(\frac{dx}{dt} = 0\), set \(-ax + bxy = 0\):
   \[x(-a + by) = 0\]
   
This equation has two nullclines:
   - \(x = 0\) (the \(x\)-nullcline)
   - \(y = \frac{a}{b}\) (the \(y\)-nullcline)

To find the nullcline for \(\frac{dy}{dt} = 0\), set \(cy - dxy = 0\):
   \[y(c - dx) = 0\]
   
This equation has two nullclines:
   - \(y = 0\) (the \(y\)-nullcline)
   - \(x = \frac{c}{d}\) (the \(x\)-nullcline)

**System 2:**
1. \(\frac{dx}{dt} = ax - \frac{ax^2}{N} - bxy\)
2. \(\frac{dy}{dt} = cy - dxy\)

To find the nullcline for \(\frac{dx}{dt} = 0\), set \(ax - \frac{ax^2}{N} - bxy = 0\):
   \[ax(1 - x/N) - bxy = 0\]
   
This equation has two nullclines:
   - \(x = 0\) (the \(x\)-nullcline)
   - \(x = N\) (the \(x\)-nullcline)

To find the nullcline for \(\frac{dy}{dt} = 0\), set \(cy - dxy = 0\):
   \[y(c - dx) = 0\]
   
This equation has two nullclines:
   - \(y = 0\) (the \(y\)-nullcline)
   - \(x = \frac{c}{d}\) (the \(x\)-nullcline)

In both systems, the nullclines are lines or curves in the \(x-y\) plane where the respective derivatives become zero. These nullclines play a crucial role in understanding the behavior of the dynamical systems and locating equilibrium points.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters a, b, c, d, and N
a = 1
b = 1
c = 1
d = 1
N = 2  # Adjust N as needed

# Define a range of x and y values
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)

# Create grids of (x, y) values
X, Y = np.meshgrid(x, y)

# Define the system of differential equations
dx_dt = a * X - (a * X**2) / N - b * X * Y
dy_dt = c * Y - d * X * Y

# Calculate the magnitude of the vectors for scaling
magnitude = np.sqrt(dx_dt**2 + dy_dt**2)

# Plot the slope fields
plt.figure(figsize=(10, 8))
plt.quiver(X, Y, dx_dt / magnitude, dy_dt / magnitude, angles='xy', cmap='autumn')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Slope Fields and Nullclines for the System')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid(True)

# Plot the nullclines
# x-nullcline
x_nullcline = a * X - (a * X**2) / N - b * X * Y
plt.contour(X, Y, x_nullcline, levels=[0], colors='blue', linestyles='dashed')

# y-nullcline
y_nullcline = c * Y - d * X * Y
plt.contour(X, Y, y_nullcline, levels=[0], colors='red', linestyles='dotted')

plt.legend(['Slope Field', 'x-Nullcline', 'y-Nullcline'])
plt.show()
```
